#### File finder

```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add


```

#### Folder creator

```python
def folder(f_name): #this function creates a folder named "attacks" in the program directory.
    try:
        if not os.path.exists(f_name):
            os.makedirs(f_name)
    except OSError:
        print ("The folder could not be created!")

folder("csvs/1")
folder("csvs/2")
```

```python
#Function that brings together disjoint csv files for ML evaluation. 
#This function also reduces the csv file size by eliminating unnecessary or unusable samples (irrelevant device names)

def csv_merger(output,csvs,devices):
    main_labels="Flow ID,Src IP,Src Port,Dst IP,Dst Port,Protocol,Timestamp,Flow Duration,Tot Fwd Pkts,Tot Bwd Pkts,TotLen Fwd Pkts,TotLen Bwd Pkts,Fwd Pkt Len Max,Fwd Pkt Len Min,Fwd Pkt Len Mean,Fwd Pkt Len Std,Bwd Pkt Len Max,Bwd Pkt Len Min,Bwd Pkt Len Mean,Bwd Pkt Len Std,Flow Byts/s,Flow Pkts/s,Flow IAT Mean,Flow IAT Std,Flow IAT Max,Flow IAT Min,Fwd IAT Tot,Fwd IAT Mean,Fwd IAT Std,Fwd IAT Max,Fwd IAT Min,Bwd IAT Tot,Bwd IAT Mean,Bwd IAT Std,Bwd IAT Max,Bwd IAT Min,Fwd PSH Flags,Bwd PSH Flags,Fwd URG Flags,Bwd URG Flags,Fwd Header Len,Bwd Header Len,Fwd Pkts/s,Bwd Pkts/s,Pkt Len Min,Pkt Len Max,Pkt Len Mean,Pkt Len Std,Pkt Len Var,FIN Flag Cnt,SYN Flag Cnt,RST Flag Cnt,PSH Flag Cnt,ACK Flag Cnt,URG Flag Cnt,CWE Flag Count,ECE Flag Cnt,Down/Up Ratio,Pkt Size Avg,Fwd Seg Size Avg,Bwd Seg Size Avg,Fwd Byts/b Avg,Fwd Pkts/b Avg,Fwd Blk Rate Avg,Bwd Byts/b Avg,Bwd Pkts/b Avg,Bwd Blk Rate Avg,Subflow Fwd Pkts,Subflow Fwd Byts,Subflow Bwd Pkts,Subflow Bwd Byts,Init Fwd Win Byts,Init Bwd Win Byts,Fwd Act Data Pkts,Fwd Seg Size Min,Active Mean,Active Std,Active Max,Active Min,Idle Mean,Idle Std,Idle Max,Idle Min,Label\n"
    ths = open(output, "w")
    ths.write(main_labels)
    for c in tqdm(csvs):
        with open(c, "r") as file:
            while True:
                line=file.readline()
                if line=="":break
                temp=line[:-1]
                temp=temp.split(",")
                if temp[-1] in devices:
                    ths.write(line)
    ths.close()
    df=pd.read_csv(output)
    print(df.groupby("Label").size())
```

```python
# find all csv files in folder 1
path="./csvs/1"
csvs=find_the_way(path,".csv")
# Create the final session 1 device group 1 file by eliminating unwanted, irrelevant samples using MAC addresses (device names)
devices=list(macs_devicegroup1.values()) #Devices names
csv_merger("session-1-DeviceGroup-1.csv",csvs,devices)
# Create the final session 1 device group 2 file by eliminating unwanted, irrelevant samples using MAC addresses (device names)
devices=list(macs_devicegroup2.values()) #Devices names
csv_merger("session-1-DeviceGroup-2.csv",csvs,devices)
```

```python
def labelcleaner(output):
    ths = open(output.replace(".csv","_.csv"), "w")

    with open(output, "r") as file:
        while True:
            line=file.readline()
            if line=="":break
            if not line.startswith("Flow"):
                #print(line)
                line=line[:-3]+"\n"
                ths.write(line)
                #print(line)
                #break
            else:ths.write(line)
    ths.close()
```
