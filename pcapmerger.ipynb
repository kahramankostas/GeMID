```python
import os
from tqdm import tqdm
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
def folder(f_name): #this function creates a folder named "attacks" in the program directory.
    try:
        if not os.path.exists(f_name):
            os.makedirs(f_name)
    except OSError:
        print ("Tthe folder could not be created!")
folder("devices")  
```

```python
def find_the_way(path,file_format,con=""):
    files_add = []
    # r=root, d=directories, f = files
    for r, d, f in os.walk(path):
        for file in f:
            if file_format in file:
                if con in file:
                    files_add.append(os.path.join(r, file))  
            
    return files_add
path="pcaps"

files_add=find_the_way('./pcaps/',".pcap")
files_add
```

```python


def create_merged(selected, name):
    selected=str(selected)
    selected=selected.replace("[","")
    selected=selected.replace("]","")
    selected=selected.replace("\'","")
    selected=selected.replace(",","")
    print(selected)

    command=f"mergecap -w {name}.pcap "+selected
    os.system(command)
```

```python
devices=[]
for i in files_add:
    temp=i.replace("\\","/")
    temp=temp.split("/")[-2]
    if temp not in devices:
        devices.append(temp)
        print(temp)
    
for i in devices:
    selected=[]
    files_add=find_the_way('./pcaps/',".pcap")
    for ii in files_add:
        if i in ii:
            selected.append(ii)
    create_merged (selected,f"./devices/{i}")       
    print(selected)
    print()
```
